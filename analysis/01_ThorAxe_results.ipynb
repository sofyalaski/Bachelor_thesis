# *ThorAxe* results

```julia
# parameters
protein = "MAPK8"
```

*ThorAxe* results for each protein is in the `benchmark` folder:

```julia
results_folder = abspath("..", "benchmark", protein)
```

All the tidy data from the homologous exons are in the `homologous_exon_table.csv` file. It's easy to work with it as a [DataFrame](http://juliadata.github.io/DataFrames.jl/stable/):

```julia
using CSV
using DataFrames
```

```julia
table_path = joinpath(results_folder, "thoraxe", "homologous_exon_table.csv")
```

```julia
const table = CSV.read(table_path);
```

```julia
show_first(data) = show(first(data, 2), allcols=true)
```

```julia
show_first(table)
```
