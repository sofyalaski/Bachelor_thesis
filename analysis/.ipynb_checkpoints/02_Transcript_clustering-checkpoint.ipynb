```julia
using Statistics

using CSV
using DataFrames
using Distances
using Clustering
using Plots
using Measures
using StatsPlots

ENV["LINES"] = 9;
```

```julia
# parameters
protein = "MAPK8"
```

```julia
function path_to_TranscriptAnnotation()
    folders = String[]
    for folder in splitpath(pwd())
        push!(folders, folder)
        if folder == "TranscriptAnnotation"
            break
        end
    end
    joinpath(folders...)
end

transcript_annotation = path_to_TranscriptAnnotation()
results_folder = joinpath(transcript_annotation, "benchmark", protein)
table_path = joinpath(results_folder, "thoraxe", "homologous_exon_table.csv")

const table = CSV.read(table_path);
```

Homologous exons sorted by their average ranking

```julia
const exons = by(table, :HomologousExon, :HomologousExonRank => mean)

sort!(exons, :HomologousExonRank_mean)
```

```julia
const exon2index = Dict(exon => i for (i, exon) in enumerate(exons[:HomologousExon]))
```

Create a binary matrix with each transcript as row and each homologous exon as a column

```julia
const transcripts = unique(table[:TranscriptIDCluster])

const transcript2index = Dict(transcript => i for (i, transcript) in enumerate(transcripts))

const n_transcripts = length(transcripts)

const n_exons = nrow(exons)

const transcript_matrix = falses(n_transcripts, n_exons)

for row in eachrow(table)
    transcript = transcript2index[row[:TranscriptIDCluster]]
    exon = exon2index[row[:HomologousExon]]
    transcript_matrix[transcript, exon] = true
end

transcript_matrix
```

Measure the Hamming distance between transcripts and perform a hierarchical clustering (`average`)

```julia
const transcript_distances = pairwise(Hamming(), transcript_matrix, dims=1)
```

```julia
const transcript_clustering = hclust(transcript_distances, linkage = :average)

plot(transcript_clustering, box=:grid)
```

```julia
const sorted_transcripts = transcripts[transcript_clustering.order]
```

```julia
const sorted_transcript_matrix = transcript_matrix[transcript_clustering.order, :]
```

```julia
const species = [ table[table[:TranscriptIDCluster] .== t, :Species][1] for t in sorted_transcripts ]
```

```julia
heatmap(sorted_transcript_matrix, 
        yflip = true,
        xrotation = 90,
        ratio = :equal,
        legend = false,
        yticks = (1:n_transcripts, species),
        xticks = (1:n_exons, exons[:HomologousExon]),
        size = (800, 800))
```

```julia
plot(
    plot(transcript_clustering, 
         box = :none, useheight = true),
    heatmap(1:n_transcripts, [1], species, 
            legend = false, box = :none),
    heatmap(sorted_transcript_matrix', 
            legend = false,
            yticks = (1:n_exons, exons[:HomologousExon])),
    layout = grid(3, 1, heights=[0.25, 0.05, 0.7]),
    margin = 0mm,
    size = (600, 600),
    xticks=nothing
)
```
